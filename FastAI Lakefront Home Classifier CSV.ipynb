```
class_name = []
outputs = []
filename = []

for i in range(len(empty_data.test_ds)):
  img = empty_data.test_ds[i][0]
  #show_image(img)
  cls_name,index,prob = learn2.predict(img)
  outputs.append(prob)
  class_name.append(cls_name)
  filename.append(str(empty_data.test_ds.x.items[i])[-13:])
```
